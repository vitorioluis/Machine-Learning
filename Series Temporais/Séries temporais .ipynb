```python
import pandas as pd
import numpy as np

%matplotlib inline
pd.set_option('display.max_columns',100)
```

```python
dados = pd.read_csv('2004-2019.tsv',sep='\t')
```

```python
dados.head()
```

```python
dados['PRODUTO'].value_counts()
```

```python
dt_diesel = dados[dados['PRODUTO'] == 'ÓLEO DIESEL'].copy()
```

```python
dt_diesel.head()
```

```python
dt_diesel['DATA FINAL'] = pd.to_datetime(dt_diesel['DATA FINAL'])
dt_diesel['DATA INICIAL'] = pd.to_datetime(dt_diesel['DATA INICIAL'])
```

```python
dt_diesel.info()
```

```python
(dt_diesel['DATA FINAL'] - dt_diesel['DATA INICIAL']).value_counts()
```

```python
diesel_treino = dt_diesel[dt_diesel['DATA INICIAL'] < '01-01-2011'].copy()
diesel_valid = dt_diesel[dt_diesel['DATA INICIAL'] >= '01-01-2011'].copy()
```

```python
diesel_treino.shape, diesel_valid.shape
```

### BASELINE

```python
y_treino = diesel_treino['PREÇO MÉDIO REVENDA']
y_valid = diesel_valid['PREÇO MÉDIO REVENDA']
```

```python
baseline_treino = diesel_treino.groupby(['ESTADO'])['PREÇO MÉDIO REVENDA'].shift(1)
baseline_valid = diesel_valid.groupby(['ESTADO'])['PREÇO MÉDIO REVENDA'].shift(1)
```

```python
diesel_treino['baseline'] = baseline_treino 
```

```python
diesel_treino
```

```python
from sklearn.metrics import mean_squared_log_error

erro = np.sqrt(
    mean_squared_log_error(
        y_valid[baseline_valid.notnull()], 
        baseline_valid[baseline_valid.notnull()]
    )
)


erro
```
